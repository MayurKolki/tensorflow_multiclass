```python
import pandas as pd
import numpy as np
from sklearn.model_selection import train_test_split
```

```python
from sklearn import preprocessing
from sklearn.metrics import confusion_matrix , classification_report , accuracy_score
```

```python
#DL - Libraray
import tensorflow as tf
import keras
from keras.models import Sequential
from keras.layers import Dense
```

```python
# Import the data 
path = 'D:/imarticus/Pythonlearn/ecoli-1.csv'
```

```python
ecoli = pd.read_csv(path)
```

```python
ecoli
```

```python
# remove the unwanted col
ecoli = ecoli.drop('sequence_name' , axis=1)
```

```python
# check the distribution of Y variabl
ecoli.lsp.value_counts()
```

```python
# Now shuffle the data
ecoli = ecoli.sample(frac=1)
```

```python
# recheck the result
ecoli.head(20)
```

```python
# Convert Y variable in to number
le = preprocessing.LabelEncoder()
target = le.fit_transform(ecoli.lsp)
ecoli['target'] = target
```

```python
ecoli[['target' , 'lsp']].head(20)
```

```python
# If we drop the Y vaiable after convert it into label encoder , it will create the difficulties to remmember thats why
#we have to use mapping to store the result

df_map = ecoli[['lsp' , 'target']].value_counts().reset_index().sort_values('target')
```

```python
df_map
```

```python
# now remove the lsp of the datasets
ecoli = ecoli.drop('lsp' , axis=1)
```

```python
# Verify the result
ecoli
```

```python
# Scale the data sets
ecoli_std = ecoli.copy()
minmax = preprocessing.MinMaxScaler()
ecoli_std.iloc[:,:] = minmax.fit_transform(ecoli.iloc[:,:])
```

```python
# Reset the Y variable
ecoli_std.target = ecoli.target
ecoli_std
```

```python
# split the data
trainx , testx , trainy , testy = train_test_split(ecoli_std.drop('target' , axis=1) , ecoli_std['target'] , test_size = 0.2)
```

```python
# Check the Y variable distribution
set(trainy)
```

```python
set(testy)
```

```python
#testy shou;d not have more that train y data
diff = set(testy)-set(trainy)

if len(diff)>0:
    print('test data has more classes than train.. resemple it')
else:
    print('Split is ok')
```

```python
# Perform tthe one hot encoding on train x and testy:
classes = len(ecoli_std.target.unique())
```

```python
# we have to find out probability of each class thats why we have to convert it into the one hot encoding.
trainy = keras.utils.to_categorical(trainy,classes)
testy = keras.utils.to_categorical(testy,classes)
```

```python
# Now build the network
inputnode = len(trainx.columns)
outputnode = classes

inputnode , outputnode
```

```python
# if unit is increase then performance increase we have to find out it ....
units = int(inputnode*outputnode + classes)
units
```

```python
# start with classifier
classifier = Sequential()
```

```python
# layer 1
classifier.add(Dense(units=units , kernel_initializer = 'uniform' , activation='relu' , input_dim = inputnode))
```

```python
#Layer 2
classifier.add(Dense(units=units , kernel_initializer = 'uniform' , activation='relu'))
```

```python
# layer 3
classifier.add(Dense(units=outputnode , kernel_initializer = 'uniform' , activation='softmax'))
```

```python
# Compile the model
classifier.compile(optimizer='adam' , metrics = ['accuracy'] , loss = 'categorical_crossentropy')
```

```python
classifier.summary()
```

```python
trainx.shape , trainy.shape
```

```python
# build a model
EPOCHS = 1000
classifier.fit(trainx , trainy , batch_size=10 , epochs=EPOCHS,validation_split=0.2)
```

```python
p1 = classifier.predict(testx)
```

```python
p2 = classifier.predict_classes(testx)
```

```python
res = pd.DataFrame({'Actual':pd.DataFrame(testy).idxmax(axis=1) , 'Predicted':predy.idxmax(axis=1)})
```
